<h1> Librerias Necesarias </h1>

```python
import pandas as pd
from sklearn import preprocessing
import pickle
```

<h1> Lectura base de datos todas las características</h1>

```python
train = pd.read_csv('train_all_feature.csv')
test = pd.read_csv('test_all_feature.csv')
```

<h1> Eliminación de características </h1>

Características que no son útiles para la clasificación

```python
train.drop(['saddr', 'sport', 'daddr', 'dport','proto','state'], axis=1, inplace=True)
test.drop(['saddr', 'sport', 'daddr', 'dport','proto','state'], axis=1, inplace=True)
```

Almacenado en el directorio características a eliminar

```python
saved_dict = pickle.load(open('./saved_dict.pkl', 'rb'))
saved_dict['feature_drop'] = ['saddr', 'sport', 'daddr', 'dport','proto','state']
pickle.dump(saved_dict, open('./saved_dict.pkl', 'wb'))
```

```python
# Tamaño de los conjuntos de datos de entrenamiento y prueba
print("Datos de entrenamiento: ", train.shape,'\nDatos de prueba: ',test.shape)
```

```python
train.columns
```

<h1> Construcción del conjunto de datos X e Y para entrenamiento y prueba </h1>

```python
# Creando los conjuntos X e Y a partir del conjunto de datos
x_train, y_train = train.drop(columns=['attack']), train['attack']
x_test, y_test = test.drop(columns=['attack']), test['attack']
```

```python
print('x_train: ',x_train.shape, 'y_train: ', y_train.shape)
print('\nx_test: ',x_test.shape, 'x_test: ', y_test.shape)
```

Almacenando en disco x_train, y_train, x_test, y_test 

```python
# Guardar todos los archivos en el disco
pickle.dump((x_train, y_train), open('final_train_all.pkl', 'wb'))
pickle.dump((x_test, y_test), open('final_test_all.pkl', 'wb'))
```

```python
# Train and Test data
x_train, y_train = pickle.load(open('final_train_all.pkl', 'rb'))
x_test, y_test = pickle.load(open('final_test_all.pkl', 'rb'))
```

<h1> Estandarizando </h1>

Estandarizando los datos de entrenamiento

```python
# Estandarizando datos de entrenamiento
scaler = preprocessing.StandardScaler()
scaler = scaler.fit(x_train[x_train.columns])

x_train[x_train.columns] = scaler.transform(x_train[x_train.columns])
```

Almacenando parámetros de estandarización de los datos de entrenamiento

```python
# Almacenado parámetros de standarización y datos de entrenamiento normalizados
pickle.dump(scaler, open('scaler_all.pkl', 'wb'))
pickle.dump((x_train, y_train), open('final_train_all_stand.pkl', 'wb'))
```

```python
x_train.describe()
```

Usando los parámetros de estantarización de los datos de entrenamiento en el conjutno de datos de prueba

```python
scaler = pickle.load(open('scaler_all.pkl', 'rb'))
x_test[x_test.columns] = scaler.transform(x_test[x_test.columns])
pickle.dump((x_test, y_test), open('final_test_all_stand.pkl', 'wb'))
```
